```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.set_option('display.max_columns' , None)
```

```python
# load dataset
data = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
data.head()
```

## Feature Engineering Step

1. Missing value

```python
# missing value for catergorical variable
feature_nan = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes == 'O']

for feature in feature_nan:
    print(feature  , np.round(data[feature].isnull().mean() , 4) , "missing %")
```

```python
def replace_nan(data , feature_nan):
    data = data.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data

data = replace_nan(data , feature_nan)
data[feature_nan].isnull().sum()
```

```python
data[feature_nan].head()
```

```python
# for numrical feature
numerical_nan = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes != 'O']

for feature in numerical_nan:
    print(feature  , np.round(data[feature].isnull().mean() , 4) , "missing %")
```

```python
for feature in numerical_nan:
    median_value = data[feature].median()
    data[feature+'nan'] = np.where(data[feature].isnull() , 1,0)
    data[feature] = data[feature].fillna(median_value)

data[feature_nan].isnull().sum()
```

```python
data.head()
```

```python
# handing datetime variable
datetime_feature = [feature for feature in data.columns if 'Yr' in feature or 'Year' in feature]
datetime_feature
```

```python
datetime_fe = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']

for feature in datetime_fe:
    if feature != 'YrSold':
        data[feature] = data['YrSold'] - data[feature]
```

```python
data[datetime_fe].head()
```

```python
# scaling numerical features
scaled_numerical = ['LotArea' , '1stFlrSF' , 'GrLivArea' , 'SalePrice']
for feature in scaled_numerical:
    data[feature] = np.log(data[feature])
```

```python
data[scaled_numerical].head()
```

## handling rare categorical variable

```python
categorical_feature = [feature for feature in data.columns if data[feature].dtypes == 'O']
```

```python
for feature in categorical_feature:
    temp = data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df = temp[temp > 0.01].index
    data[feature] = np.where(data[feature].isin(temp_df) , data[feature] ,'Rare_var')
```

```python
data.head()
```

```python
for feature in categorical_feature:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
data.head(5)
```

```python
scaling_feature=[feature for feature in data.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)
```

## Feature Scaling

```python
feature_scale=[feature for feature in data.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[feature_scale])
```

```python
scaler.transform(data[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([data[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(data[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python

data.to_csv('X_train.csv',index=False)
```
